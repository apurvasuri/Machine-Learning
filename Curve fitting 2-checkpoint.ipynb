```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,1,11)
print(x)
```

```python
t = [math.sin(i*2*math.pi) for i in x]
print(t)
```

```python
fig = plt.figure()
fig.suptitle('graph')
plt.scatter(x,t,marker='.')
```

```python
noise = np.random.normal(0,0.1,x.size)
print(noise)
```

```python
t = t+noise
print(t)
```

```python
plt.scatter(x,t,marker='.')
```

```python
n=len(x)
sumx = sum(x)
sumx2 = sum(x*x)
sumx3 = sum(x*x*x)
sumx4 = sum(x*x*x*x)
sumx5 = sum(x*x*x*x*x)
sumx6 = sum(x*x*x*x*x*x)
sumx7 = sum(x*x*x*x*x*x*x)
sumx8 = sum(x*x*x*x*x*x*x*x)
```

```python
sumt = sum(t)
sumtx = sum(t*x)
sumtx2 = sum(t*x*x)
sumtx3 = sum(t*x*x*x)
sumtx4 = sum(t*x*x*x*x)
```

```python
A = np.array([[n,sumx,sumx2,sumx3,sumx4],[sumx,sumx2,sumx3,sumx4,sumx5],[sumx2,sumx3,sumx4,sumx5,sumx6],[sumx3,sumx4,sumx5,sumx6,sumx7],[sumx4,sumx5,sumx6,sumx7,sumx8]])
B = np.array([[sumt],[sumtx],[sumtx2],[sumtx3],[sumtx4]])

W = np.matmul(np.linalg.inv(A),B);
print(W)
```

```python
t1 = [W[0][0] + W[1][0]*i+W[2][0]*i*i + W[3][0]*i*i*i + W[4][0]*i*i*i*i for i in x]
print(t1)
```

```python
plt.plot(x,t1)
plt.scatter(x,t,marker='.')
```
